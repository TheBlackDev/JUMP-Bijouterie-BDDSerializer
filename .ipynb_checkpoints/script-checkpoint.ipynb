```python
orsp = df["OR SP"]
L = [""]*len(orsp)
for i in range(len(orsp)):
    if(orsp.iloc[i] == "X"):
        L[i] += ("Or Spécial-")
orgris = df["OR GRIS"]
for i in range(len(orgris)):
    if(orgris.iloc[i] == "X"):
        L[i] += ("Or Spécial-")
oj2 = df["OR JAUNE2"]
orr = df["OR ROSE"]
o14 = df["OR 14 K"]
pl = df["PLATINE"]
ag = df["ARGENT"]
sns = df["ACIER"]

for i in range(len(oj2)):
    if(oj2.iloc[i] == "X"):
        L[i] += ("Or Jaune 2-")
for i in range(len(orr)):
    if(orr.iloc[i] == "X"):
        L[i] += ("Or Rose-")
for i in range(len(o14)):
    if(o14.iloc[i] == "X"):
        L[i] += ("Or 14K-")
for i in range(len(pl)):
    if(pl.iloc[i] == "X"):
        L[i] += ("Platine-")
for i in range(len(ag)):
    if(ag.iloc[i] == "X"):
        L[i] += ("Argent-")
for i in range(len(sns)):
    if(sns.iloc[i] == "X"):
        L[i] += ("Acier-")
        
for i in range(len(L)):
    elt = L[i]
    if(elt == "" or elt[-1] == "-"):
        L[i] = elt[:-1]
metal = df.Période.copy()
for i in range (len(L)):
    metal.iloc[i] = L[i]
metal.name = "metal"
metal.head(10)
```
